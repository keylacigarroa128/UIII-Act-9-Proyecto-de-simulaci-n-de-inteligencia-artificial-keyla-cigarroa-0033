#### Lista Carpintero

```python
# Simulador de Carpintero Inteligente con IA básica
# Desarrollado solo con Listas, If-Elif-Else y While

# Al inicio, la lista de memoria del carpintero empieza vacía
memoria_carpintero = []

print("==================================================")
print(" ¡Bienvenido al Simulador de CARPINTERO INTELIGENTE! ")
print("  Una IA básica que aprende y crea muebles ")
print("==================================================")

while True:
    # Mostramos el menú interactivo
    print("\n--- MENÚ DE COMANDOS ---")
    print("1. Mostrar inventario/memoria")
    print("2. Ordenar fabricación de un mueble")
    print("3. Reparar o barnizar un mueble")
    print("4. Resetear memoria del Carpintero")
    print("5. Apagar la IA (Salir)")

    opcion = input("\n¿Qué quieres que haga el carpintero? (Elige 1-5): ")

    # OPCIÓN 1: Ver el estado de la lista (Memoria)
    if opcion == "1":
        if memoria_carpintero == []:
            print("\n🤖 Carpintero dice: 'Mi taller está vacío. No tengo muebles en memoria.'")
        else:
            print("\n🤖 Carpintero dice: 'En mi memoria tengo registrados estos muebles:'")
            print(memoria_carpintero)

    # OPCIÓN 2: Fabricar y agregar a la lista
    elif opcion == "2":
        print("\n¿Qué mueble deseas fabricar?")
        print("- Mesa")
        print("- Silla")
        print("- Roperos")
        mueble = input("Escribe el nombre del mueble: ")

        # El "cerebro" (if-elif) decide qué madera usar según el mueble
        if mueble.lower() == "mesa":
            madera = "Madera de Roble"
            print("🤖 Carpintero: 'Analizando... Para una mesa necesito madera fuerte.'")
        elif mueble.lower() == "silla":
            madera = "Madera de Pino"
            print("🤖 Carpintero: 'Analizando... Para una silla usaré pino, es más ligera.'")
        elif mueble.lower() == "ropero":
            madera = "Madera de Caoba"
            print("🤖 Carpintero: 'Analizando... Un ropero elegante requiere caoba.'")
        else:
            madera = "Madera Común"
            print("🤖 Carpintero: 'No tengo ese mueble en mi base de datos, usaré Madera Común.'")

        # Guardamos el resultado en la lista (La IA "recuerda" lo que hizo)
        nuevo_mueble = mueble.capitalize() + " de " + madera
        memoria_carpintero = memoria_carpintero + [nuevo_mueble]
        print("🔨 [¡Mueble fabricado con éxito y guardado en la lista!] 🔨")

    # OPCIÓN 3: Interactuar con lo que ya está en la lista
    elif opcion == "3":
        if memoria_carpintero == []:
            print("\n🤖 Carpintero: 'No puedo reparar nada porque no hay muebles creados todavía.'")
        else:
            print("\nMuebles disponibles para trabajar:", memoria_carpintero)
            cual = input("¿Qué número de mueble quieres reparar/barnizar? (Empieza desde el 0): ")

            # Validación simple sin try-except
            if cual.isdigit():
                indice = int(cual)
                # Verificamos si el índice existe en la lista
                if indice < len(memoria_carpintero):
                    print("🤖 Carpintero: 'Lijando, aplicando barniz y reforzando...'")
                    memoria_carpintero[indice] = memoria_carpintero[indice] + " (¡Como Nuevo!)"
                    print("✨ ¡Mueble actualizado en la memoria!")
                else:
                    print("❌ Ese número de mueble no existe en la lista.")
            else:
                print("❌ Por favor, introduce un número válido.")

    # OPCIÓN 4: Vaciar la lista
    elif opcion == "4":
        confirmar = input("\n¿Seguro que quieres borrar la memoria del carpintero? (si/no): ")
        if confirmar.lower() == "si":
            memoria_carpintero = []
            print("🧠 Memoria formateada. Lista vacía de nuevo.")
        else:
            print("Acción cancelada.")

    # OPCIÓN 5: Romper el ciclo While
    elif opcion == "5":
        print("\n🤖 Carpintero: 'Guardando herramientas... Sistema IA Apagado. ¡Adiós!'")
        print("programa realizado por Keyla Cigarroa NL = 12")
        break

    # Validar si el usuario escribe algo que no está en el menú
    else:
        print("❌ Opción no válida. Intenta de nuevo con un número del 1 al 5.")

    print("\n--------------------------------------------------")
```
